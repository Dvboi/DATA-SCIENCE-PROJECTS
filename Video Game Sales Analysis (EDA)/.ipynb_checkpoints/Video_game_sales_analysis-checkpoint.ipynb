## TOP 10 PUBLISHERS AND DEVELOPERS (by total games)

```python
temp = df.groupby("Publisher")["Name"].count().sort_values(ascending = False)[:10]
```

```python
plt.figure(figsize = (9,3.5))
sns.barplot(y = temp.index,x = temp.values)
plt.title("TOP 10 PUBLISHERS BY TOTAL NUMBER OF GAMES")
plt.xlabel("TOTAL NUMBER OF GAMES")
plt.subplots_adjust(left = 0.25,bottom = 0.25);
```

```python
# WE SEE THERE ARE MANY SAME VIDEO-GAMES DEVELOPERS (LIKE EA,California and EA,Vegas) FOR DIFFERENT REGIONS SO WE COMBINE THEM INTO ONE AND SUM THEIR VALUES
temp = df.groupby("Developer")["Name"].count().sort_values(ascending = False)
def developer_cleaning(temp):
    temp.index = temp.index.map(lambda x: "Electronic Arts" if x[:2]=="EA" else x)
    temp.index = temp.index.map(lambda x: "Ubisoft" if "Ubisoft" in x else x)
    temp.index = temp.index.map(lambda x: "Activision" if "Activision" in x else x)
    temp.index = temp.index.map(lambda x: "Konami" if "Konami" in x else x)
    temp.index = temp.index.map(lambda x: "Omega" if "Omega" in x else x)
    temp.index = temp.index.map(lambda x: "Vicarious" if "Vicarious" in x else x)
    temp.index = temp.index.map(lambda x: "Capcom" if "Capcom" in x else x)
    temp.index = temp.index.map(lambda x: "TT Games" if "TT" in x else x)
    temp["Electronic Arts"] = temp["Electronic Arts"].sum()
    temp["Ubisoft"] = temp["Ubisoft"].sum()
    temp["Activision"] = temp["Activision"].sum()
    temp["Konami"] = temp["Konami"].sum()
    temp["Omega"] = temp["Omega"].sum()
    temp["Vicarious"] = temp["Vicarious"].sum()
    temp["Capcom"] = temp["Capcom"].sum()
    temp["TT Games"] = temp["TT Games"].sum()
    temp = temp.reset_index()
    temp = temp.drop_duplicates(subset = "Developer",keep = "first")
    return temp

temp = developer_cleaning(temp)
temp = temp[:10].reset_index(drop=True).sort_values(by = "Name",ascending = False)
temp
```

```python
plt.figure(figsize = (9,3.5))
sns.barplot(y = temp.Developer,x = temp.Name)
plt.title("TOP 10 DEVELOPERS BY TOTAL NUMBER OF GAMES")
plt.xlabel("TOTAL NUMBER OF GAMES")
plt.ylabel("")
plt.subplots_adjust(left = 0.25,bottom = 0.25);
```

### WE COMBINE DIFFERENT GENERATION GAME CONSOLES INTO ONE (EX- Ps3,Ps4,Ps5 into just Play Station)

```python
temp = df[["Platform","Year_of_Release","Global_Sales"]].dropna()
temp.Year_of_Release = temp.Year_of_Release.dt.year
# WE TAKE THE LATEST DECADE INTO ACCOUNT
temp = temp.loc[(temp["Year_of_Release"] >= 2006) & (temp["Year_of_Release"] <= 2016)]
def change(row):
    if "Wii" in row.Platform:
        return 'Wii'
    elif 'PS' in row.Platform:
        return 'Play Station'
    elif 'X' in row.Platform:
        return 'XBox'
    elif 'PC' in row.Platform:
        return 'Personal Computer'
    elif ('DS' in row.Platform) or ('GBA' in row.Platform) or ('GC' in row.Platform):
        return 'Nintendo'
    else:
        return 'DreamCast'
    
temp["Platform"] = temp.apply(change,"columns")
# temp.drop('Platform',axis = 1,inplace = True)
temp = temp[temp["Platform"]!='DreamCast']   #has really low revenue so dropped!!
temp.head()
```

```python
# different dataframes for plotting
temp = temp.groupby(["Platform","Year_of_Release"]).sum()
df_n = temp.loc["Nintendo"]
df_pc = temp.loc["Personal Computer"]
df_w = temp.loc["Wii"]
df_ps = temp.loc["Play Station"]
df_x = temp.loc["XBox"]
```

```python
plt.figure(figsize = (8,5))
plt.plot(df_x.index,df_x.Global_Sales,label = "XBox")
plt.plot(df_ps.index,df_ps.Global_Sales,"--ro",label = "Playstation")
plt.plot(df_pc.index,df_pc.Global_Sales,label = "Personal Computer")
plt.plot(df_w.index,df_w.Global_Sales,label = "Wii")
plt.plot(df_n.index,df_n.Global_Sales,label = "Nintendo",c = "cyan")
plt.legend(frameon = False,loc = 1,prop = {'size':8})
plt.title("GLOBAL CONSOLE GAME SALES OVER THE YEARS (2006-2016)")
plt.ylabel("Global Sales Revenue (in Millions of Dollars)")
plt.xlabel("TIME (in years)")
plt.xticks(np.arange(2006,2017,1));
```

##### PLAYSTATION IS THE WINNER HERE

## SALES COMPARISON BY GENRE

```python
temp = df.loc[:,["Genre","NA_Sales","EU_Sales","JP_Sales","Other_Sales"]]
temp = temp.groupby("Genre").sum()
```

```python
plt.figure()
sns.heatmap(temp,annot=True,fmt=".1f")
plt.subplots_adjust(left = 0.25)
plt.title("Sales in Millions of $");
```

```python
temp = temp.reset_index()
temp = pd.melt(temp,id_vars=["Genre"],value_vars=["NA_Sales","EU_Sales","JP_Sales","Other_Sales"],var_name="Sale_Region",
                                           value_name="Sale_Price")
```

```python
plt.figure(figsize=(8.5,6))
sns.barplot(hue=temp.Sale_Region,y=temp.Sale_Price,x=temp.Genre)
plt.legend(frameon=False)
plt.ylabel("Sale Price (in Millions of $)")
plt.xticks(rotation = 35,rotation_mode = "anchor",ha = "right");
```

## Which year had most number of game releases?

```python
temp = df.loc[:,"Year_of_Release"].dt.year
temp = temp.value_counts().reset_index()
temp.columns = ["year","count"]
final_ans = temp.sort_values(by=["year"]).reset_index(drop=True)[20:37]
final_ans.year = final_ans.year.astype(int)
final_ans.set_index("year",inplace=True)
```

```python
plt.figure()
sns.barplot(x = final_ans.index.values,y = final_ans["count"].values)
plt.xticks(rotation=35,rotation_mode="anchor",ha="right")
plt.title("GAME RELEASES OVER THE LAST 16 YRS")
plt.ylabel("NUMBER OF RELEASES")
plt.xlabel("YEAR")
plt.subplots_adjust(bottom = 0.2);
```
