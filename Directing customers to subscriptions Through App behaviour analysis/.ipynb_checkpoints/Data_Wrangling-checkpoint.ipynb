#### Feature Engineering


```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['first_open'] = data.first_open.apply(lambda x:parser.parse(x))
data['enrolled_date'] = data.enrolled_date.apply(lambda x:parser.parse(x) if isinstance(x,str) else x)
```

```python
data.dtypes
```

```python
data['difference'] = (data['enrolled_date'] - data['first_open']).astype('timedelta64[h]')
```

```python
plt.hist(data['difference'].dropna());
```

```python
plt.hist(data['difference'].dropna(),range=[0,100]);
```

##### Most differences in time are within 48 hours

```python
data.loc[data.difference>48,'enrolled'] = 0
data.drop(columns=['first_open','difference','enrolled_date'],inplace=True)
```

```python
top_screens=pd.read_csv('top_screens.csv').top_screens.values
```

```python
data['screen_list'] = data.screen_list.astype(str) + ','
```

```python
for sc in top_screens:
    data[sc]=data.screen_list.str.contains(sc).astype(int)
    data['screen_list'] = data['screen_list'].str.replace(sc+',','')
data['other'] = data.screen_list.str.count(',')
data.drop(columns=['screen_list'],inplace=True)
```

#### Funnels OR combining multiple columns into a single category

```python
savings_combined = []
credit_combined = []
cc_combined = []
loan_combined = []
for ele in data.columns:
    if 'saving' in ele.lower():
        savings_combined.append(ele)
    elif 'credit' in ele.lower():
        credit_combined.append(ele)
    elif 'cc' in ele.lower():
        cc_combined.append(ele)
    elif 'loan' in ele.lower():
        loan_combined.append(ele)
```

```python
data['savings_combined'] = data[savings_combined].sum(axis=1)
data['credit_combined'] = data[credit_combined].sum(axis=1)
data['cc_combined'] = data[cc_combined].sum(axis=1)
data['loan_combined'] = data[loan_combined].sum(axis=1)
to_drop = savings_combined + cc_combined + credit_combined + loan_combined
data.drop(columns=to_drop,inplace=True)
```
